```python
ventas_categoria = df.groupby("Product line")["Total"].sum().sort_values()

ventas_categoria.plot(kind="barh")
plt.title("Ventas por categoría")
plt.xlabel("Total de ventas")
plt.show()
```

```python
ventas_categoria = df.groupby("Product line")["Sales"].sum().sort_values()

ventas_categoria.plot(kind="barh")
plt.title("Ventas por categoría")
plt.xlabel("Total de ventas")
plt.show()
```

```python
sns.countplot(data=df, x="Payment")
plt.title("Método de pago más usado")
plt.show()
```

```python
sns.boxplot(data=df, x="Customer type", y="Sales")
plt.title("Gasto por tipo de cliente")
plt.show()
```

```python
ventas_ciudad = df.groupby("City")["Sales"].sum()

ventas_ciudad.plot(kind="bar")
plt.title("Ventas por ciudad")
plt.show()
```

El análisis exploratorio de datos permite identificar patrones relevantes en el comportamiento de las ventas. Se observa que existen diferencias en el volumen de ventas según la categoría de producto, destacándose algunas por encima de otras. Asimismo, el método de pago más utilizado refleja las preferencias de los clientes al momento de realizar sus compras. Por otro lado, el análisis del tipo de cliente muestra variaciones en el gasto total, lo cual permite distinguir comportamientos entre clientes miembros y normales. Finalmente, las ventas también presentan diferencias según la ciudad, lo que podría estar asociado a factores geográficos o de demanda.
